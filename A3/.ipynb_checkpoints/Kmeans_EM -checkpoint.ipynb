## 1.1 - K-means clustering and Expectation Maximization 
#### Implement K-means clustering algorithm using only the numpy library. You can use opencv and matplotlib libraries only to read and display images but not for clustering. Apply k-means to the provided image with K=2 and K=3. Display the resulting segmented images for 10 iterations in both cases. 

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
import math
from skimage import data, segmentation
from skimage.future import graph
from sklearn.cluster import MeanShift
```

```python
''' Plots list of images on a grid '''
def plot_figs(images, titles=None, max_col=4):
    
    # Each plot can hold up to 8 images.
    num_images = len(images)
    rows_per_set = 8 // max_col 
    fig_per_set = rows_per_set * max_col
    sets = math.ceil(num_images / fig_per_set)
    
    for s in range(sets):
        # starting index of your set
        start = s * fig_per_set
        
        # compute the number of images and rows in the current set 
        subset_len = min(start+fig_per_set, num_images) - start
        nb_rows = math.ceil(subset_len / max_col)
        
        # set subplot map 
        subplot = "{}{}".format(nb_rows, max_col)
        plt.figure(figsize=(20, 4 * nb_rows))
        for i in range(subset_len):
            plt.subplot(int("{}{}".format(subplot, i+1)))
            if titles is not None: 
                plt.title(titles[start + i])
            else:
                plt.title(str(start + i + 1))
            plt.axis('off')
            plt.imshow(images[start + i].astype(np.uint8))
    plt.show()
```

```python
''' Performs image segmentation and plots the evolution through its training process. '''
def k_means(img, k, iter=10):
    h, w, ch = img.shape
    img = img.astype(np.float64)
    
    # select the k centers for each cluster randomnly
    centers = np.random.random_sample((k, ch)) * 255.0
    evolution = []
    for it in range(iter):
        # compute distance between each pixel and cluster center
        distance = np.array([img - c for c in centers])
        distance = np.linalg.norm(distance, axis=-1)
        
        # assign pixel to closest cluster center 
        clusters = np.argmin(distance, axis=0)
        evolution.append(clusters)
        
        # compute new centers using latest assignment
        new_center = np.empty(shape=(k,ch))
        for c in range(k):
            np.mean(img[clusters == c], axis=0, out=new_center[c])
        centers = new_center
    
    # plot the evolution
    plot_figs(evolution, max_col=5)
```

```python
# read image
img = cv2.imread('home.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# show original image
plt.title('Original Image')
plt.imshow(img), plt.axis('off')
plt.show()

# do k means for 2 clusters
print("K-Means clustering with 2 clusters")
k_means(img, 2)

# do k means for 3 clusters
print("K-Means clustering with 3 clusters")
k_means(img, 3)
```

#### Implement the EM algorithm using only the numpy library. You can use opencv and matplotlib libraries only to read and display images. Apply it to the given image with 2 and 3 Gaussian components. Display the resulting segmented images for 10 iterations in both cases.

```python
''' Compute the EM algorithm and outputs the iterations '''
def compute_em(img, k, iterations=10):
        height, width, channels = img.shape
        means = np.random.random((k, channels)) * 255.0
        probability = np.full((k), (1.0 / k))
        deviation = np.full((k, channels), (255.0 / k))
        evolution = []
        for i in range(iterations):
            # E-step
            exp = np.empty(shape=(height, width, k))
            for h in range(height):
                for w in range(width): 
                    exp[h][w] = np.linalg.norm(math.e**(-0.5 * np.power((img[h][w]-means)/deviation, 2)) / (math.sqrt(2*math.pi) * deviation), axis=-1)

            denom = np.sum(exp * probability, axis=-1)
            denom = denom.reshape((height, width, 1))
            exp = exp / denom

            # draw clusters
            clusters = np.argmax(exp, axis=-1)
            evolution.append(clusters)

            # M-step 
            for c in range(k): 
                e = 0
                for h in range(height):
                    for w in range(width): 
                        e += exp[h, w, c] * img[h, w]
                means[c] = e / np.sum(exp[:,:,c])
                curr = img[clusters==c]
                deviation[c] = np.sqrt(np.var(curr, axis=0))
                probability[c] = np.size(curr) / np.size(img)

        plot_figs(evolution, max_col=5)
```

```python
print("EM Algorithm with 2 gaussian components.")
compute_em(img, 2, 10)

print("EM Algorithm with 3 gaussian components.")
compute_em(img, 3, 10)
```

#### Under what data distribution conditions would K-means and EM give the same solution, provided similar initializations?
K-means and EM would give very similar solutions in the cases when the clusters are very sparse and with lots of spacing between them. In K-means, this allows the algorithm to very easily separate the clusters. In EM, this means that there is little to no overlap between the gaussian distributions. The resulting conclusion from both cases would be boundaries that are very trivially seperable.
